# Scrapping 3000 Data

```python
from google_play_scraper import Sort, reviews, app # Mengimpor fungsi Sort, reviews, dan app dari google_play_scraper
import pandas as pd # Mengimpor pandas untuk manipulasi data
import numpy as np # Mengimpor numpy untuk operasi array

result, continuation_token = reviews(
    'com.jobstreet.jobstreet',  # ID aplikasi JobStreet
    lang='id', # bahasa Indonesia
    country='id', # negara Indonesia
    sort=Sort.NEWEST, # urutkan dari yang paling baru
    count=3000, # jumlah review yang diambil 3000
    filter_score_with=None # ambil semua review
) # Mengambil hasil review dan token lanjutan jika ada

df = pd.DataFrame(np.array(result), columns=['review']) # Mengubah hasil review menjadi DataFrame dengan kolom 'review'
df = df.join(pd.DataFrame(df.pop('review').tolist())) # Memecah kolom 'review' menjadi beberapa kolom dan menggabungkannya kembali
df.to_csv('Dataset/jobstreet_3000_selected.csv', index=False) # Menyimpan DataFrame ke file CSV
df = pd.read_csv('Dataset/jobstreet_3000_selected.csv') # Membaca kembali file CSV untuk memastikan data tersimpan dengan benar
df # Menampilkan isi DataFrame
```

# Scrapping All Data

```python
# Scrap all reviews aplikasi reku di playstore
from google_play_scraper import Sort, reviews_all

review = reviews_all(
    'com.jobstreet.jobstreet',
    sleep_milliseconds=0, # defaults to 0
    lang='id', # defaults to 'en'
    country='id', # defaults to 'us'
    sort=Sort.NEWEST # defaults to Sort.MOST_RELEVANT
)

df_jobstreet = pd.DataFrame(np.array(review), columns=['content'])
df_jobstreet = df_jobstreet.join(pd.DataFrame(df_jobstreet.pop('content').tolist()))
df_jobstreet.to_csv('Dataset/jobstreet_all_selected.csv', index=False)
```
