```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Daftar dokumen setelah stemming
docs = [
    "login",
    "masuk jobstreet",
    "info cepat pemberitahuan mengisi form lamar kerja pikir aplikasi yg lambat efisien cari kerja abu",
    "susah masuk",
    "akun log out login susah banget tolong baik akses",
    "very good app",
    "good luck for us all",
    "top",
    "ok",
    "thanks for the job"
]

# Inisialisasi TfidfVectorizer dengan hanya Term Frequency (tanpa IDF)
vectorizer = TfidfVectorizer(use_idf=False, norm=None)
X = vectorizer.fit_transform(docs)

# Ambil fitur (kata-kata unik)
features = vectorizer.get_feature_names_out()

# Hitung Term Frequency (TF)
tf_matrix = X.toarray()

# Hitung jumlah kata dalam setiap dokumen
total_terms_per_doc = tf_matrix.sum(axis=1, keepdims=True)

# Normalisasi menjadi TF (frekuensi per kata dalam dokumen)
tf_matrix = tf_matrix / total_terms_per_doc

# Buat DataFrame dalam format tabel
df_tf = pd.DataFrame(tf_matrix, columns=features, index=[f"D{i+1}" for i in range(len(docs))])

# Simpan ke CSV
df_tf.to_csv("term_frequency.csv")

# Tampilkan DataFrame
print(df_tf)
```
