```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Daftar dokumen setelah stemming
docs = [
    "login",
    "masuk jobstreet",
    "info cepat pemberitahuan mengisi form lamar kerja pikir aplikasi yg lambat efisien cari kerja abu",
    "susah masuk",
    "akun log out login susah banget tolong baik akses",
    "very good app",
    "good luck for us all",
    "top",
    "ok",
    "thanks for the job"
]

# Inisialisasi TfidfVectorizer dengan hanya Term Frequency (tanpa IDF)
vectorizer = TfidfVectorizer(use_idf=False, norm=None)
X = vectorizer.fit_transform(docs)

# Ambil fitur (kata-kata unik)
features = vectorizer.get_feature_names_out()

# Hitung Term Frequency (TF)
tf_matrix = X.toarray()

# Hitung jumlah kata dalam setiap dokumen
total_terms_per_doc = tf_matrix.sum(axis=1, keepdims=True)

# Normalisasi menjadi TF (frekuensi per kata dalam dokumen)
tf_matrix = tf_matrix / total_terms_per_doc

# Buat DataFrame dalam format tabel
df_tf = pd.DataFrame(tf_matrix, columns=features, index=[f"D{i+1}" for i in range(len(docs))])

# Simpan ke CSV
df_tf.to_csv("term_frequency.csv")

# Tampilkan DataFrame
print(df_tf)
```

```python
# Mengambil 5 kalimat pertama dan 5 kalimat terakhir dari semua dokumen
first_five = my_df['text_tokens_stemmed'].head(5)
last_five = my_df['text_tokens_stemmed'].tail(5)

# Menggabungkan kedua bagian tersebut
combined = pd.concat([first_five, last_five])

# Menghitung TF untuk setiap kalimat dalam pertimbangan positif
tf_combined = []
for doc in combined:
    words = doc.split()
    word_count = len(words)
    word_freq = {word: words.count(word) / word_count for word in words}
    tf_combined.append(word_freq)

# Menyiapkan data untuk disimpan ke CSV
tf_data_combined = []
for i, (index, doc) in enumerate(combined.items()):
    for word, freq in tf_combined[i].items():
        tf_data_combined.append({'Index': index, 'Word': word, 'TF': freq})

# Membuat DataFrame dari data TF
df_tf_combined = pd.DataFrame(tf_data_combined)

# Menyimpan DataFrame ke CSV
df_tf_combined.to_csv('sample_data/tf_data_combined.csv', index=False)

# Menampilkan DataFrame
df_tf_combined.head(10)
```

```python
import pandas as pd
import numpy as np
from collections import Counter

# Mengambil 5 kalimat pertama dan 5 kalimat terakhir
first_five = A.head(5)
last_five = A.tail(5)

# Menggabungkan kedua bagian tersebut
combined = pd.concat([first_five, last_five])

# Pastikan kita mengambil teksnya dari kolom yang benar
combined_texts2 = A.squeeze().astype(str)  # Pastikan teks berbentuk string
combined_texts = combined.squeeze().astype(str)  # Pastikan teks berbentuk string

# Jumlah total dokumen
N = 2695

# Menghitung jumlah dokumen yang mengandung setiap kata
word_doc_count = Counter()
for doc in combined_texts2:
    words = set(doc.split())  # Ambil kata unik per dokumen
    for word in words:
        word_doc_count[word] += 1  # Hitung df(t), jumlah dokumen yang mengandung kata t
        # word_doc_count[word] += 1  # Hitung df(t), jumlah dokumen yang mengandung kata t

# Menghitung IDF
idf = []
for word, df_t in word_doc_count.items():
    idf_value = np.log(N / df_t)  # IDF formula
    idf.append({'Word': word, 'IDF': idf_value, 'total jumlah dokumen': N, 'jumlah dokumen yang mengandung kata t': df_t})

# Membuat DataFrame dari data IDF
df_idf = pd.DataFrame(idf)

# Menyimpan DataFrame ke CSV
df_idf.to_csv('sample_data/idf_data.csv', index=False)

# Menampilkan DataFrame
df_idf.head(10)
```
